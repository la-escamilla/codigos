```python
import matplotlib.pyplot as plt
import numpy as np

lcdm=[-34.8,-29.5]

nodal = {4:[-36.0,-28.1],
         5:[-37.0,-28.1],
         6:[-36.5,-27.7],
         7:[-37.0,-27.6],
         8:[-37.4,-26.9],
         10:[-37.8,-26.8],
         12:[-38.7,-26.7],
         15:[-39.2,-24.6]}

x=[]
y1=[]
y2=[]
for i in nodal:
    x += [i]
    y1 += [nodal[i][0] - lcdm[0]]
    y2 += [nodal[i][1]*(2.) + lcdm[1]*(-2.)]

example = np.linspace(4,15,100)
y3=np.zeros(len(example))
plt.plot(x,y1,marker='o')
plt.plot(x,y2,marker='s')
plt.plot(example,y3,linestyle='--')
ticks = np.arange(4, 16, 1)
plt.xticks(x)
plt.xlabel('Number of nodes/bins')
plt.ylabel('Fitness')
plt.legend([r'$\ln{B_{12}}$',r'$|\Delta \chi^2|$'])
plt.savefig('graph_fitness_nodal.pdf')
```
